```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def plot_workload(ext):
    files = ["data/db_%s.csv", "data/shilldb_no_trigger_%s.csv", "data/shilldb_%s.csv", "data/shilldbctc_%s.csv"]
    
    means = []
    errs = []
    
    for f in files:
        df = pd.read_csv(f % ext)
        means.append(df.mean()[1])
        errs.append(df.std()[1])
                       
    plt.errorbar([1, 2, 3, 4], means, yerr=errs, fmt='o')
    plt.xlim(0, 5)
    plt.show()
    return map(lambda x : (x -  means[0]) / means[0] * 100, means)
```

```python
plot_workload("read_revised")
```

```python
plot_workload("mixed")
```

```python
plot_workload("write")
```

```python
plot_workload("insert")
```
